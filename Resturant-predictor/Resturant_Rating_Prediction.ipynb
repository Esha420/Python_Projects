```python
x = df[['Average Cost for two', 'Has Table booking', 'Has Online delivery', 'Price range']]
```

```python
y = df[['Aggregate rating']]
```

```python
x
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
#Here 1 is yes and 0 is no.
x['Has Table booking'] = le.fit_transform(x['Has Table booking'])
x['Has Online delivery'] = le.fit_transform(x['Has Online delivery'])
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def modelresults(predictions):
    print("Mean abslute error on model is {}".format(mean_absolute_error(y_test,predictions)))
    print("Root mean square error {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
    
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
predictionsfromlr = lr.predict(X_test)
```

```python
modelresults(predictionsfromlr)
```

```python
predictionsfromlr
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C" : [0.001,0.01,0.1,0.5], "kernel" : ["linear", "rbf", "ploy"],"degree" : [2,3,4]}
```

```python
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
gridsvr.fit(X_train, y_train)
```

```python
predsgridsvr = gridsvr.predict(X_test)
```

```python
modelresults(predsgridsvr)
```

```python
predsgridsvr = gridsvr.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {
    "max_depth" : [10,30],
    "min_samples_leaf" : [1,2],
    "min_samples_split" : [2,5]
}
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = treemodel, param_grid = param_grid)
```

```python
grid_tree.fit(X_train,y_train)
```

```python
treepredictions = grid_tree.predict(X_test)
```

```python
modelresults(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {"max_depth": [5,15], "n_estimators" : [2,5,10]}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(X_train,y_train)
```

```python
randomforestpredictions = gridrfr.predict(X_test)
```

```python
modelresults(randomforestpredictions)
```

```python
gridrfr.best_params_
```

```python
grid_tree.best_params_

```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn_param_grid = {"n_neighbors": [49,51,53,55,65]}
knn_grid_search = GridSearchCV(knn, knn_param_grid)
```
